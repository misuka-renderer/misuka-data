```python
import mitsuba as mi
mi.set_variant('cuda_acoustic', 'llvm_acoustic', 'scalar_acoustic')
import pyfar as pf
import matplotlib.pyplot as plt
```

```python
import numpy as np

sampling_rate = 1000.0

# Load the two EXR files
etc_ref = np.array(mi.TensorXf(mi.Bitmap("test_shoebox_primal_ref.exr")))

# Convert to pyfar Signals (each column is a frequency band channel)
sig_ref = pf.Signal(etc_ref.T, sampling_rate)

# Plot both
fig, axes = plt.subplots(1, 1, sharex=True, figsize=(10, 6))
pf.plot.time(sig_ref, dB=True, ax=axes, label="ref")
plt.tight_layout()
plt.show()
```
